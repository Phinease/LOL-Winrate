# Model Selection

```python
# import all needed models
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import LocalOutlierFactor
from sklearn.decomposition import PCA
```

```python
from sklearn.feature_selection import RFE
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Lasso
from sklearn.linear_model import ElasticNet
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.ensemble import RandomForestRegressor
```

```python
main = pd.read_csv("test_10mins.csv")
X = main.copy().drop(columns = ['blueWins', 'gameId'])
Y = main.copy()['blueWins']
X_train, X_test, Y_train, Y_test = train_test_split (X, Y, test_size = 0.20, random_state = 1)

pipelines = []

pipelines.append(('LR', Pipeline([('Scaler', StandardScaler()),('LR',LinearRegression())])))
pipelines.append(('LASSO', Pipeline([('Scaler', StandardScaler()),('LASSO', Lasso())])))
pipelines.append(('EN', Pipeline([('Scaler', StandardScaler()),('EN', ElasticNet())])))
pipelines.append(('KNN', Pipeline([('Scaler', StandardScaler()),('KNN', KNeighborsRegressor())])))
pipelines.append(('CART', Pipeline([('Scaler', StandardScaler()),('CART', DecisionTreeRegressor())])))
pipelines.append(('GBR', Pipeline([('Scaler', StandardScaler()),('GBR', GradientBoostingRegressor())])))
pipelines.append(('RFR', Pipeline([('Scaler', StandardScaler()),('RFR', RandomForestRegressor())])))

results = []
names = []
for score in ['neg_mean_squared_error', 'neg_mean_absolute_error', 'r2']:
    print("\nTest", score, ": ")
    for name, model in pipelines:
        kfold = KFold(n_splits = 10, shuffle = True, random_state = 1)
        cv_result = cross_val_score(model, X_train, Y_train, cv = kfold, scoring = score)
        results.append(cv_result)
        names.append(name)
        msg = "%s:\t%f +/- (%f)" % (name, cv_result.mean(), cv_result.std())
        print(msg)
```

```python
pipe = Pipeline([('PCA', PCA(n_components = 22)),('GBR',GradientBoostingRegressor(n_estimators = 50, loss = 'ls'))])
pipe.fit(X_train, Y_train)
Y_pred = pipe.predict(X_test)
truth = np.array(Y_test)
for i in range(20):
    print(truth[i], '-', Y_pred[i])
```

# Pre-processing

```python
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import LocalOutlierFactor
from sklearn.decomposition import PCA
```

```python
pipelines = []
pipelines.append(('GBR\t', Pipeline([('Scaler', StandardScaler()),('GBR', GradientBoostingRegressor())])))

for i in range(20, 30):
    pipelines.append(('GBR - PCA - ' + str(i), Pipeline([('PCA', PCA(n_components = i)),('GBR', GradientBoostingRegressor())])))

results = []
names = []
for name, model in pipelines:
    kfold = KFold(n_splits = 10, shuffle = True, random_state = 1)
    cv_result = cross_val_score(model, X_train, Y_train, cv = kfold, scoring = 'neg_mean_squared_error')
    results.append(cv_result)
    names.append(name)
    msg = "%s:\t%f +/- (%f)" % (name, cv_result.mean(), cv_result.std())
    print(msg)
```

# Hyper Parameters Optimizing

```python
from sklearn.model_selection import GridSearchCV
```

```python
pca = PCA()
gbr = GradientBoostingRegressor()
pipe = Pipeline(steps=[('pca', pca), ('gbr', gbr)])

param_grid = {
    'pca__n_components': range(20, 25),
    'gbr__n_estimators': [50, 100, 150],
    'gbr__learning_rate': [0.1, 0.3, 0.01],
    'gbr__loss': ['ls', 'lad', 'huber', 'quantile']
}

search = GridSearchCV(pipe, param_grid, n_jobs = -1)
search.fit(X_train, Y_train)
print("Best parameter (CV score=%0.3f):" % search.best_score_)
print(search.best_params_)
```

```python
pipe = Pipeline([('PCA', PCA(n_components = 22)),('GBR',GradientBoostingRegressor(n_estimators = 50, loss = 'ls'))])
kfold = KFold(n_splits = 10, shuffle = True, random_state = 1)
cv_result = cross_val_score(pipe, X_train, Y_train, cv = kfold, scoring = 'r2')
msg = "Best:\t%f +/- (%f)" % (cv_result.mean(), cv_result.std())
print(msg)
```

```python
pca = PCA()
lr = LinearRegression()
pipe = Pipeline(steps=[('pca', pca), ('lr', lr)])

param_grid = {
    'pca__n_components': range(20, 25)
}

search = GridSearchCV(pipe, param_grid, n_jobs = -1)
search.fit(X_train, Y_train)
print("Best parameter (CV score=%0.3f):" % search.best_score_)
print(search.best_params_)
```
